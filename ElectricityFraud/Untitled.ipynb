```python
import pandas as pd
import numpy as np
```

```python
label_df = pd.read_csv("client_train.csv")
```

```python
target_good = 4
num_bad_candidates = label_df["target"].value_counts()[1.0]
num_good_candidates = target_good * num_bad_candidates
```

```python
bad_clients = label_df[label_df["target"]==1.0]["client_id"].tolist()
good_clients = label_df[label_df["target"]==0.0]["client_id"].tolist()
```

```python
import random
rnd_seed = 35
random.Random(rnd_seed).shuffle(good_clients)
```

```python
short_good_clients = good_clients[:num_good_candidates]
```

```python
clients = short_good_clients + bad_clients
targets = [0]*num_good_candidates + [1]*num_bad_candidates
```

```python
clients_df = pd.DataFrame({"client": clients, "target": targets})
clients_df = clients_df.sort_values("client")
clients_df.to_csv("targets.csv", index = False)
```

```python
chunk_size = 1000000
reading_iters = 5
data_df = pd.DataFrame()
for i in range(0, reading_iters):
    if i == 0:
        read_df = pd.read_csv("invoice_train.csv", header = 0, nrows = chunk_size )
    else:
        read_df = pd.read_csv("invoice_train.csv", header = 0, nrows = chunk_size, skiprows = range(1, chunk_size * i) )
    filter_df = read_df[read_df["client_id"].isin(clients)]
    data_df = pd.concat([data_df, filter_df])
```
